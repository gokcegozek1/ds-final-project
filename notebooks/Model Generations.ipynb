```python
import fal_client
import pandas as pd
from urllib.request import urlretrieve
from glob import glob
import base64
import os

# Set your API Key
os.environ["FAL_KEY"] = "f56b2fe4-f2dd-4753-a772-62928483b48c:5d679712b244e5b9ab3e4534f01f8823"

# Load product item Excel
product_df = pd.read_excel('/Users/gokcegozek/Desktop/Product Items.xlsx')

# Sort model images in ascending order
model_image_folder = '/Users/gokcegozek/Desktop/model_images/passed level 1'
model_images = sorted(glob(os.path.join(model_image_folder, "*.png")))

# Create output folder
output_folder = '/Users/gokcegozek/Desktop/generated_item_fittings'
os.makedirs(output_folder, exist_ok=True)

# CSV record list
records = []

# Load previously generated files to skip
existing_files = set(os.listdir(output_folder))

# Function to convert image to base64
def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# Loop through all model image x product image combinations
for i, row in product_df.iterrows():
    product_type = row.get('Product Type', 'Product')
    color = row.get('Color', 'Unknown')
    price = row.get('Product Price', '0')
    image_url = row.get('Image Model URL')

    # Download clothing image temporarily
    clothing_img_path = f'/tmp/clothing_{i}.png'
    urlretrieve(image_url, clothing_img_path)

    # Convert clothing image to base64
    clothing_image_base64 = image_to_base64(clothing_img_path)

    for model_img_path in model_images:
        model_filename = os.path.basename(model_img_path)

        # Construct final filename
        final_filename = f"{product_type}_{i}_{color}_{price}_{model_filename}"
        final_output_path = os.path.join(output_folder, final_filename)

        if final_filename in existing_files:
            print(f"⏭️ Skipping already generated: {final_filename}")
            continue

        # Convert model image to base64
        model_image_base64 = image_to_base64(model_img_path)

        print(f"🚀 Generating: {final_filename}")

        try:
            # Submit to fal.ai
            response = fal_client.subscribe(
                "fashn/tryon",
                arguments={
                    "model_image": f"data:image/png;base64,{model_image_base64}",
                    "garment_image": f"data:image/png;base64,{clothing_image_base64}",
                    "category": "one-pieces"
                },
                with_logs=True
            )

            # Retrieve generation result
            result = response.get("images", [])
            if result:
                image_url = result[0].get("url")
                if image_url:
                    urlretrieve(image_url, final_output_path)

                    # Append to CSV record
                    record = row.to_dict()
                    record["Model Image Name"] = model_filename
                    record["Generated File Name"] = final_filename
                    record["Generated Image URL"] = image_url
                    records.append(record)

        except Exception as e:
            print(f"❌ Failed: {final_filename} - {e}")
            with open(os.path.join(output_folder, "failed_generations.txt"), "a") as f:
                f.write(final_filename + "\n")
            continue

# Save CSV with all records
records_df = pd.DataFrame(records)
records_df.to_csv('/Users/gokcegozek/Desktop/generated_item_fittings/generated_item_fittings_records.csv', index=False)

print("✅ All items generated and saved successfully.")
```
